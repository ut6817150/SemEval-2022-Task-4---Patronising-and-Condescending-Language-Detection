````python
import re, json, random, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from sklearn.metrics import (
    f1_score, precision_score, recall_score,
    classification_report, confusion_matrix, ConfusionMatrixDisplay
)
from transformers import (
    AutoTokenizer, AutoModelForSequenceClassification,
    TrainingArguments, Trainer, DataCollatorWithPadding
)
from torch.utils.data import Dataset
import shutil
from pathlib import Path

warnings.filterwarnings('ignore')

cuda_available = torch.cuda.is_available()
mps_available  = torch.backends.mps.is_available()
DEVICE = torch.device('cuda' if cuda_available else ('mps' if mps_available else 'cpu'))
print(f'Device: {DEVICE}')
````

````python
# ── Shared helpers (self-contained — no imports from 2_training.ipynb) ──────────────

_URL_RE    = re.compile(r'https?://\S+|www\.\S+')
_EMAIL_RE  = re.compile(r'\b[\w\.-]+@[\w\.-]+\.\w+\b')
_HANDLE_RE = re.compile(r'@\w+')
_WS_RE     = re.compile(r'\s+')

_HTML_ENTITIES = {
    '&amp;': '&',  '&lt;': '<',   '&gt;': '>',
    '&quot;': '"', '&apos;': "'", '&#39;': "'",
    '&nbsp;': ' ', '&ldquo;': '"', '&rdquo;': '"',
    '&lsquo;': "'", '&rsquo;': "'",
}

def preprocess(text: str) -> str:
    """
    Matches 2_training.ipynb preprocessing exactly.
    HTML entity decoding → URL/email/handle replacement → whitespace collapse.
    """
    if not isinstance(text, str):
        return ''
    for ent, char in _HTML_ENTITIES.items():
        text = text.replace(ent, char)
    text = re.sub(r'&#\d+;', ' ', text)
    text = _URL_RE.sub(' url ', text)
    text = _EMAIL_RE.sub(' email ', text)
    text = _HANDLE_RE.sub(' user ', text)
    return _WS_RE.sub(' ', text).strip()


class PCLDataset(Dataset):
    """
    Matches 2_training.ipynb PCLDataset exactly.
    Tokenizes without padding so DataCollatorWithPadding pads per-batch dynamically.
    """
    def __init__(self, texts, labels, tokenizer, max_length=256):
        self.encodings = tokenizer(
            texts, truncation=True, max_length=max_length,
            padding=False, return_tensors=None
        )
        self.labels = labels

    def __len__(self): return len(self.labels)

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item


def get_probs(model_dir, texts, max_length=256):
    """Load a single saved model and return P(PCL) for each text."""    
    tokenizer = AutoTokenizer.from_pretrained(model_dir)
    model     = AutoModelForSequenceClassification.from_pretrained(model_dir).to(DEVICE)
    model.eval()
    dataset  = PCLDataset(texts, [0]*len(texts), tokenizer, max_length)
    collator = DataCollatorWithPadding(tokenizer=tokenizer)
    output   = Trainer(
        model=model,
        args=TrainingArguments(output_dir='./tmp_infer', report_to='none'),
        data_collator=collator,
    ).predict(dataset)
    probs = torch.softmax(
        torch.tensor(output.predictions, dtype=torch.float32), dim=-1
    )[:, 1].numpy()
    del model
    if cuda_available: torch.cuda.empty_cache()
    return probs


def ensemble_inference(models_dir, texts, threshold=None, max_length=256):
    """
    Load all models from models_dir, average probabilities, apply threshold.
    Returns (predictions, avg_probs, per_model_probs, threshold).
    """
    if threshold is None:
        with open(os.path.join(models_dir, 'threshold.json')) as f:
            threshold = json.load(f)['best_threshold']
        print(f"Threshold: {threshold:.2f}  (from threshold.json)")

    model_dirs = sorted([
        os.path.join(models_dir, d) for d in os.listdir(models_dir)
        if os.path.isdir(os.path.join(models_dir, d))
        and os.path.exists(os.path.join(models_dir, d, 'config.json'))
    ])
    print(f"Models: {[os.path.basename(d) for d in model_dirs]}")

    per_model_probs = []
    for d in model_dirs:
        print(f"  Loading {os.path.basename(d)} ...", end=' ', flush=True)
        p = get_probs(d, texts, max_length)
        per_model_probs.append(p)
        print(f"mean P(PCL)={p.mean():.3f}")

    avg_probs   = np.mean(np.stack(per_model_probs, axis=0), axis=0)
    predictions = (avg_probs >= threshold).astype(int)
    print(f"Ensemble: {predictions.sum()} / {len(predictions)} predicted PCL")
    return predictions, avg_probs, per_model_probs, threshold


print('Helpers ready.')
````

## Part A — Global Evaluation

Fresh independent data loads, ensemble inference, write `dev.txt` and `test.txt`.

### A.1 Load Dev Set (fresh)

````python
# Fresh load — completely independent of 2_training.ipynb runtime variables.
# skiprows=4 skips the 3-line disclaimer + blank line at the top of the training TSV.
_data = pd.read_csv(
    PCL_TSV_PATH, sep='\t', skiprows=4, header=None,
    names=['par_id', 'art_id', 'keyword', 'country', 'text', 'orig_label'],
    quoting=3, dtype={'par_id': str}
)
_data = _data.dropna(subset=['orig_label']).reset_index(drop=True)
_data['text'] = _data['text'].fillna('')
_data['label'] = (_data['orig_label'].astype(int) >= 2).astype(int)

_deids = pd.read_csv(DEV_IDS_PATH, usecols=['par_id'], dtype={'par_id': str})
dev_df = _deids.merge(
    _data[['par_id', 'text', 'label', 'keyword', 'country']], on='par_id', how='inner'
).reset_index(drop=True)
dev_df['orig_text']   = dev_df['text']
dev_df['text']        = dev_df['text'].apply(preprocess)
dev_df['text_length'] = dev_df['text'].str.split().str.len()

print(f"Dev set : {len(dev_df):,} rows | PCL: {dev_df.label.sum()} | Non-PCL: {(dev_df.label==0).sum()}")
````

### A.2 Load Test Set (fresh)

````python
# The test TSV has NO header rows — data begins on line 1.
# No skiprows here, 3,832 lines.

if not os.path.exists(TEST_TSV_PATH):
    print(f"Test file not found at {TEST_TSV_PATH} — test.txt will be skipped.")
    test_df = None
else:
    test_df = pd.read_csv(
        TEST_TSV_PATH, sep='\t', header=None,
        names=['par_id', 'art_id', 'keyword', 'country', 'text'],
        quoting=3, dtype={'par_id': str}
    )
    test_df = test_df.dropna(subset=['text']).reset_index(drop=True)
    test_df['text'] = test_df['text'].apply(preprocess)
    print(f"Test set : {len(test_df):,} rows")
    print(f"  Expected 3,832 rows — match: {len(test_df) == 3832}")
````

### A.3 Baseline Inference

````python
baseline_tokenizer = AutoTokenizer.from_pretrained(BASELINE_DIR)
baseline_model     = AutoModelForSequenceClassification.from_pretrained(BASELINE_DIR).to(DEVICE)
baseline_model.eval()

baseline_output = Trainer(
    model=baseline_model,
    args=TrainingArguments(output_dir='./tmp_eval', report_to='none'),
    data_collator=DataCollatorWithPadding(tokenizer=baseline_tokenizer),
).predict(PCLDataset(dev_df['text'].tolist(), dev_df['label'].tolist(), baseline_tokenizer))

baseline_probs = torch.softmax(
    torch.tensor(baseline_output.predictions, dtype=torch.float32), dim=-1
)[:, 1].numpy()
baseline_preds = (baseline_probs >= 0.5).astype(int)
baseline_f1    = f1_score(dev_df['label'], baseline_preds, pos_label=1, average='binary')

print(f"Baseline Dev F1 (PCL) : {baseline_f1:.4f}")

# clean up workspace
del baseline_model
if cuda_available: torch.cuda.empty_cache()

tmp_eval_path = Path("./tmp_eval")

if tmp_eval_path.exists():
    shutil.rmtree(tmp_eval_path)
    print(f"Deleted: {tmp_eval_path.resolve()}")
else:
    print(f"Not found (nothing to delete): {tmp_eval_path.resolve()}")
````

### A.4 BestModel Inference — Dev Set

````python
print("Running BestModel ensemble on dev set...\n")
dev_preds, dev_avg_probs, dev_per_model_probs, THRESHOLD = ensemble_inference(
    models_dir = BESTMODEL_DIR,
    texts      = dev_df['text'].tolist(),
)

dev_f1 = f1_score(dev_df['label'], dev_preds, zero_division=0)
print(f"\nBestModel Dev F1 (PCL) : {dev_f1:.4f}")

# clean up workspace
tmp_infer_path = Path("./tmp_infer")

if tmp_infer_path.exists():
    shutil.rmtree(tmp_infer_path)
    print(f"Deleted: {tmp_infer_path.resolve()}")
else:
    print(f"Not found (nothing to delete): {tmp_infer_path.resolve()}")
````

### A.5 BestModel Inference — Test Set

````python
if test_df is not None:
    print("Running BestModel ensemble on test set...\n")
    test_preds, test_avg_probs, _, _ = ensemble_inference(
        models_dir = BESTMODEL_DIR,
        texts      = test_df['text'].tolist(),
    )
else:
    test_preds = None
    print("Test set not available — skipping.")

# clean up workspace
tmp_infer_path = Path("./tmp_infer")

if tmp_infer_path.exists():
    shutil.rmtree(tmp_infer_path)
    print(f"Deleted: {tmp_infer_path.resolve()}")
else:
    print(f"Not found (nothing to delete): {tmp_infer_path.resolve()}")
````

### A.6 Write dev.txt and test.txt

````python
# Both files written to root directory
with open('./dev.txt', 'w') as f:
    f.write('\n'.join(map(str, dev_preds.tolist())))
print(f"dev.txt  : {len(dev_preds)} lines  |  PCL F1 = {dev_f1:.4f}")
print(f"  Outperforms baseline (0.48) : {dev_f1 > 0.48}")

if test_preds is not None:
    with open('./test.txt', 'w') as f:
        f.write('\n'.join(map(str, test_preds.tolist())))
    print(f"\ntest.txt : {len(test_preds)} lines  |  PCL predicted: {test_preds.sum()} ({test_preds.mean():.3f})")
    print(f"  Threshold : {THRESHOLD:.2f}")
else:
    print("\ntest.txt skipped — test set not found.")
````

## Part B — Local Evaluation

### B.1 Error Analysis

Split the dev set into four outcome groups and analyse where each model succeeds or fails.

````python
dev_df = dev_df.copy()
dev_df['true']         = dev_df['label'].values
dev_df['best_pred']    = dev_preds
dev_df['base_pred']    = baseline_preds
dev_df['best_correct'] = (dev_df['best_pred'] == dev_df['true'])
dev_df['base_correct'] = (dev_df['base_pred'] == dev_df['true'])

def error_group(row):
    if row['best_correct'] and row['base_correct']:         return 'Both correct'
    if not row['best_correct'] and not row['base_correct']: return 'Both wrong'
    if row['best_correct']:                                 return 'BestModel only'
    return 'Baseline only'

dev_df['error_group'] = dev_df.apply(error_group, axis=1)

fn_df = dev_df[(dev_df['true'] == 1) & (dev_df['best_pred'] == 0)]  # missed PCL
fp_df = dev_df[(dev_df['true'] == 0) & (dev_df['best_pred'] == 1)]  # wrongly flagged
tp_df = dev_df[(dev_df['true'] == 1) & (dev_df['best_pred'] == 1)]  # correctly detected

total_pcl     = dev_df['true'].sum()
total_non_pcl = (dev_df['true'] == 0).sum()

print("Error group distribution:")
for g, c in dev_df['error_group'].value_counts().items():
    print(f"  {g:<25} : {c:>4}  ({c/len(dev_df)*100:.1f}%)")
print()
print(f"False Negatives (missed PCL)  : {len(fn_df):>3}  ({len(fn_df)/total_pcl*100:.1f}% of all PCL)")
print(f"False Positives (wrong flag)  : {len(fp_df):>3}  ({len(fp_df)/total_non_pcl*100:.1f}% of all Non-PCL)")
print(f"True Positives  (correct PCL) : {len(tp_df):>3}  ({len(tp_df)/total_pcl*100:.1f}% of all PCL)")
````

#### B.1.1 Error Breakdown by Keyword and Text Length

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
BLUE   = "#4C72B0"
ORANGE = "#DD8452"

# Build per-keyword stats for both models
kw_stats = []
for kw, grp in dev_df.groupby('keyword'):
    pcl = (grp['true'] == 1).sum()
    non_pcl = (grp['true'] == 0).sum()
    best_fn = ((grp['true']==1) & (grp['best_pred']==0)).sum()
    best_fp = ((grp['true']==0) & (grp['best_pred']==1)).sum()
    base_fn = ((grp['true']==1) & (grp['base_pred']==0)).sum()
    base_fp = ((grp['true']==0) & (grp['base_pred']==1)).sum()
    kw_stats.append({
        'keyword': kw, 'pcl_total': pcl, 'non_pcl_total': non_pcl,
        'best_fn': best_fn, 'best_fp': best_fp,
        'base_fn': base_fn, 'base_fp': base_fp,
        'best_fn_rate': best_fn / pcl       if pcl > 0     else 0,
        'best_fp_rate': best_fp / non_pcl   if non_pcl > 0 else 0,
        'base_fn_rate': base_fn / pcl       if pcl > 0     else 0,
        'base_fp_rate': base_fp / non_pcl   if non_pcl > 0 else 0,
    })
kw_df = pd.DataFrame(kw_stats)

# Panel 1: FN rate by keyword — Baseline vs BestModel
kw_fn = kw_df.sort_values('best_fn_rate', ascending=True)
y = np.arange(len(kw_fn))
w = 0.35
axes[0].barh(y + w/2, kw_fn['base_fn_rate'] * 100, w, label='Baseline', color=BLUE, alpha=0.75)
axes[0].barh(y - w/2, kw_fn['best_fn_rate'] * 100, w, label='BestModel', color=ORANGE, alpha=0.75)
axes[0].set_yticks(y)
axes[0].set_yticklabels(kw_fn['keyword'])
axes[0].set_xlabel('False Negative Rate (%)')
axes[0].set_title('FN Rate by Keyword')
axes[0].legend(fontsize=9)
axes[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

# Panel 2: FP rate by keyword — Baseline vs BestModel
kw_fp = kw_df.sort_values('best_fp_rate', ascending=True)
y2 = np.arange(len(kw_fp))
axes[1].barh(y2 + w/2, kw_fp['base_fp_rate'] * 100, w, label='Baseline', color=BLUE, alpha=0.75)
axes[1].barh(y2 - w/2, kw_fp['best_fp_rate'] * 100, w, label='BestModel', color=ORANGE, alpha=0.75)
axes[1].set_yticks(y2)
axes[1].set_yticklabels(kw_fp['keyword'])
axes[1].set_xlabel('False Positive Rate (%)')
axes[1].set_title('FP Rate by Keyword')
axes[1].legend(fontsize=9)
axes[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

plt.tight_layout(); plt.show()

# Summary table
print("\nPer-keyword error rates — Baseline vs BestModel:\n")
summary = kw_df[['keyword', 'pcl_total', 'non_pcl_total',
                  'base_fn_rate', 'best_fn_rate',
                  'base_fp_rate', 'best_fp_rate']].copy()
summary['base_fn_rate'] = (summary['base_fn_rate'] * 100).round(1).astype(str) + '%'
summary['best_fn_rate'] = (summary['best_fn_rate'] * 100).round(1).astype(str) + '%'
summary['base_fp_rate'] = (summary['base_fp_rate'] * 100).round(1).astype(str) + '%'
summary['best_fp_rate'] = (summary['best_fp_rate'] * 100).round(1).astype(str) + '%'
summary.columns = ['Keyword', 'PCL', 'Non-PCL',
                    'Base FN Rate', 'Best FN Rate',
                    'Base FP Rate', 'Best FP Rate']
summary = summary.sort_values('Best FN Rate', ascending=False)
print(summary.to_string(index=False))
````

#### B.1.2 Concrete Examples

````python
# Add ensemble and baseline probability to error dataframes
fn_df_conf = fn_df.copy()
fp_df_conf = fp_df.copy()
fn_df_conf['prob'] = dev_avg_probs[fn_df.index]
fp_df_conf['prob'] = dev_avg_probs[fp_df.index]
fn_df_conf['base_prob'] = baseline_probs[fn_df.index]
fp_df_conf['base_prob'] = baseline_probs[fp_df.index]

# Most confident FNs: PCL the model was most sure was Non-PCL
fn_sorted = fn_df_conf.sort_values('prob')        # lowest P(PCL) first
# Most confident FPs: Non-PCL the model was most sure was PCL
fp_sorted = fp_df_conf.sort_values('prob', ascending=False)  # highest P(PCL) first

# PCL caught by baseline but missed by BestModel
baseline_catches = fn_df_conf[fn_df_conf['base_pred'] == 1].sort_values('prob')

print("=" * 70)
print("FALSE NEGATIVES — PCL the model was most confident about missing")
print("(sorted by P(PCL) ascending — most wrong first)")
print("=" * 70)
for i, (_, row) in enumerate(fn_sorted.head(10).iterrows(), 1):
    print(f"\n[FN {i}] keyword={row['keyword']}  |  BestModel P(PCL)={row['prob']:.3f}  |  Baseline P(PCL)={row['base_prob']:.3f}")
    print(f"  {row['orig_text'][:300]}")

print()
print("=" * 70)
print("FALSE POSITIVES — Non-PCL the model was most confident about flagging")
print("(sorted by P(PCL) descending — most wrong first)")
print("=" * 70)
for i, (_, row) in enumerate(fp_sorted.head(10).iterrows(), 1):
    print(f"\n[FP {i}] keyword={row['keyword']}  |  BestModel P(PCL)={row['prob']:.3f}  |  Baseline P(PCL)={row['base_prob']:.3f}")
    print(f"  {row['orig_text'][:300]}")

print()
print("=" * 70)
print(f"BASELINE CATCHES MISSED BY BESTMODEL — {len(baseline_catches)} cases")
print("(PCL the ensemble lost by moving to a higher-precision operating point)")
print("=" * 70)
for i, (_, row) in enumerate(baseline_catches.head(10).iterrows(), 1):
    print(f"\n[BC {i}] keyword={row['keyword']}  |  BestModel P(PCL)={row['prob']:.3f}  |  Baseline P(PCL)={row['base_prob']:.3f}")
    print(f"  {row['orig_text'][:300]}")
````

#### B.1.3 Where BestModel Corrects Baseline and Vice Versa

````python
best_only = dev_df[dev_df['error_group'] == 'BestModel only']
base_only = dev_df[dev_df['error_group'] == 'Baseline only']

print(f"Cases BestModel gets right, Baseline gets wrong : {len(best_only)}")
print(f"Cases Baseline gets right, BestModel gets wrong : {len(base_only)}")
print(f"Net gain from BestModel                         : +{len(best_only) - len(base_only)}")
print()
print("BestModel-only correct — by true label:")
print(best_only['true'].value_counts().rename({0:'Non-PCL', 1:'PCL'}).to_string())
print()
print("Baseline-only correct — by true label:")
print(base_only['true'].value_counts().rename({0:'Non-PCL', 1:'PCL'}).to_string())
````

### B.2 Confusion Matrices

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for ax, preds, title in [
    (axes[0], baseline_preds, f'Baseline RoBERTa-base\nDev F1={baseline_f1:.3f}'),
    (axes[1], dev_preds,      f'BestModel Ensemble\nDev F1={dev_f1:.3f}'),
]:
    cm = confusion_matrix(dev_df['true'], preds)
    ConfusionMatrixDisplay(cm, display_labels=['Non-PCL', 'PCL']).plot(ax=ax, colorbar=False, cmap='Blues')
    tn, fp_c, fn_c, tp_c = cm.ravel()
    ax.set_title(title, fontsize=12)
    ax.set_xlabel(f'Predicted\nTPR={tp_c/(tp_c+fn_c):.2f}  FPR={fp_c/(fp_c+tn):.2f}', fontsize=10)

plt.suptitle('Confusion Matrices — Dev Set', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.show()
````

### B.3 Ensemble Ablation Study

````python
# dev_per_model_probs already collected during A.4 — no reloading needed
abl = []
for n in range(1, len(dev_per_model_probs) + 1):
    avg = np.mean(np.stack(dev_per_model_probs[:n], axis=0), axis=0)
    preds_n = (avg >= THRESHOLD).astype(int)
    seed, pw, _ = ENSEMBLE_CONFIGS[n-1]
    abl.append({
        'n_models': n,
        'last_added': f'seed={seed} w={pw}',
        'f1':        f1_score(dev_df['true'], preds_n, zero_division=0),
        'precision': precision_score(dev_df['true'], preds_n, zero_division=0),
        'recall':    recall_score(dev_df['true'], preds_n, zero_division=0),
    })
abl_df = pd.DataFrame(abl)

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(abl_df['n_models'], abl_df['f1'],        marker='o', lw=2,   label='F1',       color='#4C72B0')
ax.plot(abl_df['n_models'], abl_df['precision'],  marker='s', lw=1.5, label='Precision', color='#DD8452', linestyle='--')
ax.plot(abl_df['n_models'], abl_df['recall'],     marker='^', lw=1.5, label='Recall',    color='#55A868', linestyle=':')
ax.axhline(baseline_f1, linestyle='-.', color='grey', lw=1, label=f'Baseline ({baseline_f1:.3f})')
ax.set_xlabel('Number of models in ensemble'); ax.set_ylabel('Score')
ax.set_xticks(abl_df['n_models'])
ax.set_xticklabels([f"{r['n_models']}\n({r['last_added']})" for _, r in abl_df.iterrows()], fontsize=8)
ax.set_title('Ensemble Ablation: F1 as Models Are Added')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()

print(abl_df[['n_models','last_added','f1','precision','recall']].to_string(index=False))
````

### B.4 Weight Sensitivity Analysis

````python
w_results = []
for (seed, pos_weight, _), probs in zip(ENSEMBLE_CONFIGS, dev_per_model_probs):
    preds_bt = (probs >= THRESHOLD).astype(int)
    preds_05 = (probs >= 0.5).astype(int)
    w_results.append({
        'pos_weight': pos_weight, 'seed': seed,
        f'F1@{THRESHOLD:.2f}': f1_score(dev_df['true'], preds_bt, zero_division=0),
        'F1@0.5':              f1_score(dev_df['true'], preds_05, zero_division=0),
        'Precision':           precision_score(dev_df['true'], preds_bt, zero_division=0),
        'Recall':              recall_score(dev_df['true'], preds_bt, zero_division=0),
    })
w_df = pd.DataFrame(w_results)

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].plot(w_df['pos_weight'], w_df[f'F1@{THRESHOLD:.2f}'], marker='o', lw=2, color='#4C72B0', label=f'F1@{THRESHOLD:.2f}')
axes[0].plot(w_df['pos_weight'], w_df['F1@0.5'], marker='s', lw=1.5, linestyle='--', color='#4C72B0', alpha=0.5, label='F1@0.5')
axes[0].axhline(baseline_f1, linestyle='-.', color='grey', lw=1, label=f'Baseline ({baseline_f1:.3f})')
axes[0].set_xlabel('Positive class weight'); axes[0].set_ylabel('F1 (PCL)')
axes[0].set_title('Per-Model F1 by pos_weight'); axes[0].legend(); axes[0].grid(True, alpha=0.3)

sc = axes[1].scatter(w_df['Recall'], w_df['Precision'],
                     c=w_df['pos_weight'], cmap='RdYlGn_r', s=130,
                     edgecolors='black', linewidths=0.5, zorder=5)
for _, row in w_df.iterrows():
    axes[1].annotate(f"w={row['pos_weight']}", (row['Recall'], row['Precision']),
                     textcoords='offset points', xytext=(6, 4), fontsize=9)
plt.colorbar(sc, ax=axes[1], label='pos_weight')
axes[1].set_xlabel('Recall'); axes[1].set_ylabel('Precision')
axes[1].set_title('Precision–Recall Tradeoff by Weight')
axes[1].grid(True, alpha=0.3)

plt.tight_layout(); plt.show()

print("\nWeight sensitivity table:")
print(w_df.to_string(index=False))
print("\nLower weight = more conservative (high precision). Higher weight = more aggressive (high recall).")
print("The ensemble averages across all four operating points.")
````
